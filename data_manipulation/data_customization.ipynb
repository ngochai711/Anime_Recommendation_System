```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
data_location = './../dataset/'

df_anime = pd.read_csv(os.path.join(data_location, 'anime.csv'))
df_anime_with_synopsis = pd.read_csv(os.path.join(data_location, 'anime_with_synopsis.csv'))
df_animelist = pd.read_csv(os.path.join(data_location, 'animelist.csv'))
df_rating_complete = pd.read_csv(os.path.join(data_location, 'rating_complete.csv'))
df_watching_status = pd.read_csv(os.path.join(data_location, 'watching_status.csv'))
```

```python
db_anime = df_anime.set_index('MAL_ID') \
   .join(df_anime_with_synopsis.set_index('MAL_ID').drop(['Name', 'Genres', 'Score'], axis=1)) \
   .drop(['Score-10', 'Score-9', 'Score-8', 'Score-7', 'Score-6', 'Score-5', 'Score-4', 'Score-3', 'Score-2', 'Score-1', 'Members', 'Favorites', 'Watching', 'Completed',
         'On-Hold', 'Dropped', 'Plan to Watch', 'Ranked'], axis=1)
   
db_anime
```

```python
db_rating = pd.DataFrame(data=df_anime['Rating'].unique().reshape(-1,1), columns=['Rating'])
db_rating.index += 1

db_rating
```

```python
db_anime['Rating'] = db_anime['Rating'].apply(lambda x: db_rating[db_rating['Rating'] == x].index[0])
db_anime = db_anime.replace({"Unknown": np.nan})

db_anime
```

```python
db_anime.to_csv("../dataset/db_anime.csv")
db_rating.to_csv("../dataset/db_rating.csv")
```
